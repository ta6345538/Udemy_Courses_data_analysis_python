## Which Courses have a higher number of subscribers free or paid?

```python
sns.barplot(x='is_paid',y='num_subscribers', data=df)
```

## Which level has the highest number of subscribers?

```python
sns.barplot(x='level',y='num_subscribers',data=df)
plt.xticks(rotation=60)
plt.show()
```

## Most Popular course title

```python
df[df['num_subscribers'].max()==df['num_subscribers']]['course_title']
```

## 10 Most popular courses as per number of subscribers

```python
top_10=df.sort_values(by='num_subscribers',ascending=False).head(10)
```

```python
sns.barplot(x='num_subscribers',y='course_title',data=top_10)
plt.show()
```

## The course which is having the highest number of reviews

```python
plt.figure(figsize=(10,4))
sns.barplot(x='subject',y='num_reviews',data=df)
```

## Does price affect number of reviews?

```python
plt.figure(figsize=(15,6))
sns.scatterplot(x='price',y='num_reviews',data=df)
plt.show()
```

## Total number of courses related to Python

```python
print("Total number of courses related to Python =",len(df[df['course_title'].str.contains('python',case=False)]))
```

## 10 Most popular 'Python' courses as per number of subscribers

```python
top_python=df[df['course_title'].str.contains('python',case=False)].sort_values(by='num_subscribers',ascending=False).head(10)
```

```python
sns.barplot(x='num_subscribers',y='course_title',data=top_python)
plt.show()
```
